```python
import numpy as np
import pandas as pd
import sqlite3
```

```python
covidDeaths = pd.read_excel(
    'CovidDeaths.xlsx', 
    header=0)

covidVaccinations = pd.read_excel(
    'CovidVaccinations.xlsx', 
    header=0)
```

```python
covidDeaths.head()
```

```python
print(covidDeaths.columns)
covidVaccinations.columns
```

```python
DBcovidDeaths = covidDeaths[['continent','location','date','total_cases','new_cases','total_deaths','new_deaths',
                             'icu_patients','hosp_patients']]
DBcovidDeaths.head()
```

```python
DBcovidVaccinations = covidVaccinations [['continent', 'location', 'date', 'population', 'total_tests', 'new_tests', 
                                          'positive_rate', 'total_vaccinations', 'people_vaccinated',
                                         'people_fully_vaccinated', 'total_boosters', 'new_vaccinations']]

DBcovidVaccinations.head()
```

```python
db_conn = sqlite3.connect("Covid.db")
c = db_conn.cursor()
```

```python
c.execute(
    """
    CREATE TABLE DBcovidDeaths (
        continent TEXT,
        location TEXT,
        date TEXT,
        total_cases REAL,
        new_cases REAL,
        total_deaths REAL,
        new_deaths REAL,
        icu_patients REAL,
        hosp_patients REAL,        
        PRIMARY KEY(location, date),
        FOREIGN KEY(location) REFERENCES DBcovidVaccinations(location),
        FOREIGN KEY(date) REFERENCES DBcovidVaccinations(date)
        );
     """
)


c.execute(
    """
    CREATE TABLE DBcovidVaccinations (
        continent TEXT,
        location TEXT,
        date TEXT,
        population REAL,
        total_tests REAL,
        new_tests REAL,
        positive_rate INTEGER,
        total_vaccinations REAL,
        people_vaccinated REAL,
        people_fully_vaccinated REAL,
        total_boosters REAL,
        new_vaccinations REAL,
        PRIMARY KEY(location, date),
        FOREIGN KEY(location) REFERENCES DBcovidDeaths(location),
        FOREIGN KEY(date) REFERENCES DBcovidDeaths(date)
        );
     """
)
```

```python
DBcovidDeaths.to_sql('DBcovidDeaths', db_conn, if_exists='append', index=False)
DBcovidVaccinations.to_sql('DBcovidVaccinations', db_conn, if_exists='append', index=False)
```
